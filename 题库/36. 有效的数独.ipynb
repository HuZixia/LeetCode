```python
# 遍历，三个集合

class Solution(object):
    def isValidSudoku(self, board):
        rows = [set() for _ in range(9)]
        clos = [set() for _ in range(9)]
        boxes = [set() for _ in range(9)]
        
        for i in range(9):
            for j in range(9):
                num = board[i][j]
                if num == '.':
                    continue
                box_index = (i // 3) * 3 + j // 3
                if num in rows[i] or num in clos[j] or num in boxes[box_index]:
                    return False
                rows[i].add(num)
                clos[j].add(num)
                boxes[box_index].add(num)
        return True

solution = Solution()
board= [["5","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]  
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]  
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
print(solution.isValidSudoku(board))



board= [["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]  
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]              
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]  
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
print(solution.isValidSudoku(board))
```
