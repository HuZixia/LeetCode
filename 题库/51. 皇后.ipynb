```python
# 递归 + 回溯

class Solution(object):
    def solveNQuens(self, n):
        
        # 创建棋盘，将状态转换成字符串列表
        def create_board(state):
            board = []
            for row in state:
                # 将每一行连接成一个字符串，加入到棋盘中
                board.append(''.join(row))
            return board
        
        # 判断在 row col 放置皇后是否有效
        def is_valid(state, row, col):
            for i in range(row):
                # 检查同一列中是否有皇后
                if state[i][col] == 'Q':
                    return False
                # 检查左上对角线是否有皇后
                if col - (row - i) >= 0 and state[i][col - (row -i)] == 'Q':
                    return False
                # 检查右上对角线是否有皇后
                if col + (row - i) < len(state) and state[i][col + (row - i)] == 'Q':
                    return False
            return True
        
        # 回溯算法解决问题
        def backtrack(state, row):
            # 如果当前行等于棋盘大小，说明已经找到了一种解法，加入到结果中
            if row == len(state):
                # 将当前状态加入结果中
                result.append(create_board(state))
                return 
            # 遍历当前行的每一列，尝试放置皇后
            for col in range(len(state)):
                if is_valid(state, row, col):
                    state[row][col] = 'Q'
                    # 递归处理下一行
                    backtrack(state, row + 1)
                    # 回溯，撤销当前放置的皇后
                    state[row][col] = '.'
        
        result = []
        state = [['.'] * n for _ in range(n)]
        backtrack(state, 0)
        return result
        
solution = Solution()
print(solution.solveNQuens(4)) 
print(solution.solveNQuens(1))
print(solution.solveNQuens(9))
```
